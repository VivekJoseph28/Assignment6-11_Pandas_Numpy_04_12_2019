### Problem Statement 1:
1) How-to-count-distance-to-the-previous-zero
For each value, count the difference of the distance from the previous zero (or the start of the Series, whichever is closer) and if there are no previous zeros, print the position
Consider a DataFrame df where there is an integer column {'X'[7, 2, 0, 3, 4, 2, 5, 0, 3, 4]}
The values should therefore be [1, 2, 0, 1, 2, 3, 4, 0, 1, 2]. Make this a new column 'Y'.
import pandas as pd
df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})

```python
import pandas as pd

## Input Series
s = pd.Series([7, 2, 0, 3, 4, 2, 5, 0, 3, 4])

## Group by and mask method to compute the distance from the Zero
print((s.groupby(s.eq(0).cumsum().mask(s.eq(0))).cumcount() + 1).mask(s.eq(0), 0).tolist())
```

## OR

```python
## Input list 
l = [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]
i = 0
r = []
## For loop to compute the distance from zero for the list
for element in l:
    if element != 0:
        i += 1
    else:
        i = 0
    r.append(i)
print (r)
```

### 2) Create a DatetimeIndex that contains each business day of 2015 and use it to index a Series of random numbers.

```python
import datetime
import numpy as np
import pandas as pd
```

```python
x=pd.date_range(start='2015-01-01',end='2015-12-30',freq='B')
len(x)
```

```python
S=np.random.randint(0,100,len(x))
```

```python
Sales_dist = pd.DataFrame(S,index=x,columns=['Sales'])
print(Sales_dist)
```

### 3) Find the sum of the values in s for every Wednesday

```python
S3=Sales_dist.groupby(x.weekday_name)['Sales'].sum().loc['Wednesday']
round(S3,2)
```

### 4) Average For each calendar month

```python
S2=Sales_dist.groupby(pd.Grouper(freq='M'))['Sales'].mean()
round(S2,2)
```

### 5) For each group of four consecutive calendar months in s, find the date on which the highest value occurred.

```python
S6=Sales_dist.groupby(pd.Grouper(freq='Q'))['Sales'].idxmax(axis = 1)
S7=Sales_dist.groupby(pd.Grouper(freq='Q'))['Sales'].max()
print(S6.to_string(index=False),S7.to_string(index=False))
```

#### 4) What is the median name occurrence in the dataset

```python
US1=US_Baby.loc[:,['Name','Count']].groupby('Name').sum()
US1.sort_values("Count", axis = 0, ascending =False, 
                 inplace = True, na_position ='last') 
US2=US1[US1.Count==np.median(US1.Count)].reset_index()
US2.sort_values("Name", axis = 0, ascending =True, 
                 inplace = True, na_position ='last') 
US2
```
